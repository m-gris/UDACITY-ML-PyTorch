## Load the data

Here you'll use `torchvision` to load the data ([documentation](http://pytorch.org/docs/0.3.0/torchvision/index.html)). The data should be included alongside this notebook, otherwise you can [download it here](https://s3.amazonaws.com/content.udacity-data.com/nd089/flower_data.tar.gz). The dataset is split into three parts, training, validation, and testing. For the training, you'll want to apply transformations such as random scaling, cropping, and flipping. This will help the network generalize leading to better performance. You'll also need to make sure the input data is resized to 224x224 pixels as required by the pre-trained networks.

The validation and testing sets are used to measure the model's performance on data it hasn't seen yet. For this you don't want any scaling or rotation transformations, but you'll need to resize then crop the images to the appropriate size.

The pre-trained networks you'll use were trained on the ImageNet dataset where each color channel was normalized separately. For all three sets you'll need to normalize the means and standard deviations of the images to what the network expects. For the means, it's `[0.485, 0.456, 0.406]` and for the standard deviations `[0.229, 0.224, 0.225]`, calculated from the ImageNet images.  These values will shift each color channel to be centered at 0 and range from -1 to 1.
 

````python
data_dir = 'flowers'
train_dir = data_dir + '/train'
valid_dir = data_dir + '/valid'
test_dir = data_dir + '/test'
````

````python
# TODO: Define your transforms for the training, validation, and testing sets

mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

train_transform = transforms.Compose([transforms.RandomRotation(25),
                                       transforms.RandomResizedCrop(224),
                                       transforms.RandomHorizontalFlip(),
                                       transforms.ToTensor(),
                                       transforms.Normalize(mean, std)])


valid_transform = transforms.Compose([ transforms.Resize(255),
                                        transforms.CenterCrop(224),
                                        transforms.ToTensor(), 
                                        transforms.Normalize(mean, std)
                                      ])


test_transform = transforms.Compose([ transforms.Resize(255),
                                        transforms.CenterCrop(224),
                                        transforms.ToTensor(), 
                                        transforms.Normalize(mean, std)
                                      ])

# TODO: Load the datasets with ImageFolder
train_datasets = datasets.ImageFolder(train_dir, transform = train_transform)
valid_datasets = datasets.ImageFolder(valid_dir, transform = valid_transform)
test_datasets  = datasets.ImageFolder(test_dir, transform = test_transform)

# TODO: Using the image datasets and the trainforms, define the dataloaders
batch_size = 64

train_loaders = torch.utils.data.DataLoader(train_datasets, batch_size, shuffle = True)
valid_loaders = torch.utils.data.DataLoader(valid_datasets, batch_size)
test_loaders = torch.utils.data.DataLoader(test_datasets, batch_size)
````

````python
# Freeze parameters 
for parameter in model.parameters():
    parameter.requires_grad = False
````

````python
# Define a new, untrained feed-forward network as a classifier, 
## using ReLU activations and dropout

flower_categories = 102
dropout = 0.35


# densenet121
in_features=1024

classifier = nn.Sequential(
                           # INPUT LAYER
                           nn.Linear(in_features, 512), 
                           nn.ReLU(), 
                           nn.Dropout(dropout),  
                           
                           # HIDDEN LAYER 1
                           nn.Linear(512, 256), 
                           nn.ReLU(), 
                           nn.Dropout(dropout), 
    
    
                           # OUTPUT LAYER
                           nn.Linear(256, flower_categories),                        
                           nn.LogSoftmax(dim =1)
                           )

##########################################################################################
# FOR .vgg16
# in_features = 25088
# classifier = nn.Sequential(
#                            # INPUT LAYER
#                            nn.Linear(in_features, 8192), 
#                            nn.ReLU(), 
#                            nn.Dropout(dropout),  
                           

#                            # HIDDEN LAYER 1
#                            nn.Linear(8192, 2048), 
#                            nn.ReLU(), 
#                            nn.Dropout(dropout), 
    
        
#                            # HIDDEN LAYER 2
#                            nn.Linear(2048, 512), 
#                            nn.ReLU(), 
#                            nn.Dropout(dropout), 

    
#                            # OUTPUT LAYER
#                            nn.Linear(512, flower_categories),                        
#                            nn.LogSoftmax(dim =1)
#                            )

##########################################################################################

# Replace the pre-trained network classifier with our own
model.classifier = classifier


# Instantiate the loss function
criterion = nn.NLLLoss()

# Instantiate the optimizer
optimizer = optim.Adam(model.classifier.parameters(), lr = 0.003)
````

````python
# Dynamic device selection 
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

model.to(device)

# JUST TO CHEK
device
````

##### TRAINING

````python
epochs = 10
steps = 0

running_loss = 0

validate_every = 25

training_losses = []
validation_losses = []
validation_accuracies = []


for epoch in range(epochs):
    
    ####### TRAINING #######
    for images, labels in train_loaders:
        
        steps += 1
        
        # Move input and label tensors to the default device
        images, labels = images.to(device), labels.to(device)
        
        # CLEAR THE GRADIENTS
        optimizer.zero_grad()
        
        log_probs = model.forward(images) # FORWARD PASS
        loss = criterion(log_probs, labels) # COMPUTE LOSS
        loss.backward() # BACK PROP
        optimizer.step() # UPDATE WEIGHTS

        running_loss += loss.item() # ACCUMULATE LOSS
        
     
    
        if steps % validate_every == 0:
            
            ####### VALIDATING #######
            test_loss = 0
            accuracy = 0
            
            # TURN OFF DROPOUT DURING VALIDATION
            model.eval()
            
            # TURN OFF GRADIENTS - COMPUTING EFFICIENCY
            with torch.no_grad():
                for images, labels in valid_loaders:
                    # Move input and label tensors to the default device
                    images, labels = images.to(device), labels.to(device)
                   
                    log_probs = model.forward(images) # FORWARD PASS
                    batch_loss = criterion(log_probs, labels) # COMPUTE LOSS
                    test_loss += batch_loss.item() # ACCUMULATE LOSS
                    
                    # Calculate accuracy
                    probs = torch.exp(log_probs)
                    # Get highest prob. prediction
                    top_p, top_class = probs.topk(1, dim=1)
                    # Compare predictions to labels
                    equals = top_class == labels.view(*top_class.shape)
                    # Compute accuracy
                    accuracy += torch.mean(equals.type(torch.FloatTensor)).item()
            
            
            
            # TRACKING LOSSES & # TRACK ACCURACY
            training_loss = running_loss/validate_every
            validation_loss = test_loss/len(valid_loaders)
            validation_accuracy = accuracy/len(valid_loaders)
            
            training_losses.append(training_loss)
            validation_losses.append(validation_loss)
            validation_accuracies.append(validation_accuracy)  
            
            # PRINTING
            print(f"Epoch {epoch+1} out of {epochs}, step number {steps}\n"
                  f"Training loss: {training_loss:.3f}.. "
                  f"Validation loss: {validation_loss:.3f}.. "
                  f"Validation accuracy: {validation_accuracy*100:.5f}%")
            print()
                       
                      
            # RESET RUNNING LOSS
            running_loss = 0
            
            # SWITCH DROPOUT BACK ON FOR TRAINING
            model.train()
````

````python
# PLOTTING LOSSES
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

loss_accu_df = pd.DataFrame({'Training Losses':training_losses, 
                             'Validation Losses':validation_losses, 
                             'Validation Accuracies':validation_accuracies})


fig, ax = plt.subplots(ncols = 2, figsize = (14, 5))

ax[0].plot(loss_accu_df['Training Losses'], label = 'Training Losses')
ax[0].plot(loss_accu_df['Validation Losses'], label = 'Validation Losses')
ax[0].legend()


ax[1].plot(loss_accu_df['Validation Accuracies'], label = 'Validation Accuracies')

ax[1].legend();
````
